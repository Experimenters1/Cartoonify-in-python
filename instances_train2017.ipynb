Bạn có thể tải file chú thích **instances_train2017.json** từ bộ dữ liệu COCO **(Common Objects in Context)** bằng cách sử dụng **script Python** sau. Script này sẽ tự động tải xuống và giải nén file chú thích:

```python
import os
import requests
import zipfile
```

```python
from google.colab import drive
drive.mount('/content/data')
```

```python
# Đường dẫn nơi file zip sẽ được tải xuống
download_path = '/content/data/MyDrive/DoAntot/code/instances_train2017.json/download'  # Thay đổi đường dẫn này theo nhu cầu của bạn

# Đường dẫn nơi file zip sẽ được giải nén
extract_path = '/content/data/MyDrive/DoAntot/code/instances_train2017.json/extract'  # Thay đổi đường dẫn này theo nhu cầu của bạn
```

```python
# Tạo các thư mục nếu chúng không tồn tại
if not os.path.exists(download_path):
    os.makedirs(download_path)

if not os.path.exists(extract_path):
    os.makedirs(extract_path)
```

```python
# URL for COCO 2017 train annotations
url = 'http://images.cocodataset.org/annotations/annotations_trainval2017.zip'

# Tên file zip sau khi tải về
local_zip_file = os.path.join(download_path, 'annotations_trainval2017.zip')
```

```python
# Downloading the annotations
print('Downloading COCO 2017 train annotations...')
response = requests.get(url, stream=True)
with open(local_zip_file, 'wb') as file:
    for chunk in response.iter_content(chunk_size=128):
        file.write(chunk)
```

```python
print('Download complete. Extracting files...')

# Extracting the zip file
with zipfile.ZipFile(local_zip_file, 'r') as zip_ref:
    zip_ref.extractall(extract_path)  # Extract to the specified directory

```

```python
print('Download complete. Extracting files...')

# Extracting the zip file
with zipfile.ZipFile(local_zip_file, 'r') as zip_ref:
    zip_ref.extractall(extract_path)  # Extract to the specified directory

print('Extraction complete.')
```
